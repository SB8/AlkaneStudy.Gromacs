```julia
"""function read_xvg(filename::String)
    
    # For now, just ignore comment lines (#) and labels (@)
    
    open(filename) do fileIO
        for line in eachline(fileIO)
            if line[1] ∉ ['#', '@'])
                lindat = parse.(Float64, String.(split(line)))
            end
        end
        
    end
    
    # Return array
    return 0
end"""


function read_xvg(filename::String)
   
    open(filename, "r") do fileIO
        # Remove hash comments
        xy = readdlm(fileIO, comments=true, comment_char='#')
        # Remove @ (label/legend) lines for now
        i = 0
        for r in eachrow(xy)
            i += (r[1][1] == '@')
        end
    
        return Float64.(xy[i+1:end,1:2])
    end
end

```

```julia
# TraPPE ---------------------------------------------------------------------
T1 = 340; T2 = T1-100

simdir = "TraPPE_1024xC16_anneal_$(T1)-$(T2)-$(T1)K_100ns"
xvg1 = "density_$(T1)-$(T2)K.xvg"
xvg2 = "density_$(T2)-$(T1)K.xvg"

density1 = read_xvg(joinpath("..", simdir, xvg1))
density2 = read_xvg(joinpath("..", simdir, xvg2))

# Shift second density trajectory (to follow from first)
density2[:,1] .+= density1[end,1]

plot(density1[:,1], density1[:,2], 
    xaxis = ("Time (ps)"), 
    yaxis = ("Density (kg/m³)"),
    label="$(T1)-$(T2)K")
plot!(density2[:,1], density2[:,2], label="$(T2)-$(T1)K")
```
